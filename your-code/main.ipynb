```python
# SUM STEPS BY MONTH_ENCODED - WE LOOK AT THE SUM, ALTHOUGH THEY HAVE NOT REQUESTED

# First let´s group the data sets for the relevant Group By´s.
# We will first look at MONTHLY ANALYSIS

fitbit_monthly_df = fitbit_df[['Months', 'Steps']].groupby('Months', as_index=False).sum()

fitbit_monthly_df = fitbit_monthly_df.sort_values(by="Steps")

fitbit_monthly_df
```

```python
# MEDIAN STEPS BY MONTH_ENCODED

# First let´s group the data sets for the relevant Group By´s.
# We will first look at MONTHLY ANALYSIS

fitbit_monthly_df = fitbit_df[['Months', 'Steps']].groupby('Months', as_index=False).median()

fitbit_monthly_df = fitbit_monthly_df.sort_values(by="Steps")

fitbit_monthly_df
```

```python
# Let´s now plot the data

x = fitbit_monthly_df['Months']
y = fitbit_monthly_df['Steps']


plt.figure(figsize=(12,4))
plt.bar(x,y, label='# of Steps', color = 'cyan')
plt.title('Steps by month', fontsize=20)
plt.xlabel('Month', fontsize=16)
plt.ylabel('Number of steps', fontsize=16)
plt.legend()

#plt.grid(True)
plt.grid(axis='y')
plt.savefig('bar_plot_props_built.png', dpi = 400, quality=100)
plt.show()
plt.clf()
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND

# Secondly, let´s group the data sets for the relevant Group By´s.
# We will first look at WORK/WEEKEND ANALYSIS

fitbit_work_df = fitbit_df[['Work_or_Weekend', 'Steps']].groupby('Work_or_Weekend', as_index=False).median()

fitbit_work_df
```

```python
# Let´s now plot the data

x = fitbit_work_df['Work_or_Weekend']
y = fitbit_work_df['Steps']


plt.figure(figsize=(12,4))
plt.bar(x,y, label='# of Steps', color = 'cyan')
plt.title('Steps during Work of Weekend', fontsize=20)
plt.xlabel('Month', fontsize=16)
plt.ylabel('Number of steps', fontsize=16)
plt.legend()

#plt.grid(True)
plt.grid(axis='y')
plt.savefig('bar_plot_props_built.png', dpi = 400, quality=100)
plt.show()
plt.clf()
```
